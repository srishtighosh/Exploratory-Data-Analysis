##  Data Handling and EDA

> #### calendar Analysis

```python
calendar['date'] = pd.to_datetime(calendar['date'], format = "%Y-%m-%d")
calendar['day'] = calendar['date'].dt.day
calendar['month'] = calendar['date'].dt.month
calendar['Year'] = calendar['date'].dt.year
calendar['day_name_full'] = calendar['date'].dt.strftime('%A')
calendar['month_name_full'] = calendar['date'].dt.strftime('%B')
```

```python
calendar['available'].replace('t','1',inplace=True)
calendar['available'].replace('f','0', inplace=True)
```

```python
calendar.head()
```

```python
# count of Listing based on each month.

a = calendar.groupby(['month_name_full'])['listing_id'].count().sort_values(ascending=False).to_frame().reset_index()
a = a.rename(columns = {'month_name_full' : 'Months','listing_id' : 'Number of Listings'})
a
```

```python
from matplotlib import pyplot as plt

# we will use autopct to indicate percent lables on pie chart

plt.pie(a['Number of Listings'], labels = a['Months'], autopct = '%.2f%%', pctdistance = 0.6,radius=3)

plt.rcParams['font.size'] = 15

plt.show()
```

```python
# How many listings were published in every Year.

b = calendar.groupby(['Year'])['listing_id'].count().sort_values(ascending=False).to_frame().reset_index()
b = b.rename(columns = {'listing_id' : 'Number of Listings'})
b
```

```python
from matplotlib import pyplot as plt

plt.pie(b['Number of Listings'], labels = b['Year'], autopct = '%.2f%%', pctdistance = 0.6,radius=3)

plt.rcParams['font.size'] = 15

plt.show()
```

```python
# How many listings were published in every Year and in each month.

c = calendar.groupby(['Year','month_name_full'])['listing_id'].count().to_frame().reset_index()
c = c.rename(columns = {'month_name_full' : 'Month','listing_id' : 'Number of Listings'})
#c1 = c.drop(['Number of Listings'],axis=1)
c
```

```python
import seaborn as sns
sns.countplot(y="month_name_full",hue="Year", data=calendar)
```

> #### Listings and Listings_brief Analysis

```python
pd.set_option('display.max_columns', None)
```

```python
listing_merge = pd.merge(listings,listings_brief,left_on = "id",right_on = "id")
listing_merge.head(2)
```

```python
# Top 5 host where the maximum no. of listings are published
d = listing_merge.groupby(['host_name_x'])['id'].count().sort_values(ascending=False).to_frame().reset_index().head(10)
d = d.rename(columns = {'host_name_x' : 'Host Name','id' : 'Count'})
d
```

```python
sns.barplot("Count", y="Host Name", data=d ,palette="Blues_d")
plt.title("Top 10 Host")
```

```python
# How many location are incorrect
f = listing_merge['is_location_exact'].value_counts().to_frame().reset_index()
f['index']=f['index'].replace('f','No').replace('t', 'Yes')
f
```

```python

sns.barplot("is_location_exact", y="index", data=f ,palette="muted")
plt.ylabel('Exact Location (Yes/No)')
plt.xlabel('Count')
plt.title("Exact Location (Yes/No)")
```

```python
# Which property type have how many listings
g = listing_merge.groupby(['property_type'])['id'].count().sort_values().to_frame().reset_index()
g
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.barplot("id", y="property_type", data=g ,palette="muted")
plt.ylabel('Count')
plt.xlabel('Property Type')
plt.title("Property wise Listing")
```

```python
# How many listings for each Room Type 

h = listing_merge.groupby(['room_type_x'])['id'].count().to_frame().reset_index()
h
```

```python
sns.set(rc={'figure.figsize':(7,5)})

sns.barplot("id", y="room_type_x", data=h ,palette="BuGn_r")
plt.ylabel('Count')
plt.xlabel('Room Type')
plt.title("Room wise Listing")
```

```python
# How many can accomodate in each property

sns.set(rc={'figure.figsize':(12,9)})

sns.barplot("accommodates", y="property_type", data=listing_merge ,palette="muted")
plt.title("How many can accomodate in each property")
```

```python
# How many bathrooms in each property.

sns.set(rc={'figure.figsize':(12,9)})

sns.barplot("bathrooms", y="property_type", data=listing_merge ,palette="dark")
plt.title("bathrooms in each property")
```

```python
# How many bedrooms in each property.

sns.set(rc={'figure.figsize':(12,9)})

sns.barplot("bedrooms", y="property_type", data=listing_merge ,palette="colorblind")
plt.title("bedrooms in each property")
```

```python
# Average price (in Dollars) of each property 

j = listing_merge.groupby(['property_type'])['price_y'].mean().sort_values(ascending=False).reset_index()
j
```

```python
sns.barplot("price_y", y="property_type", data=j, palette="Blues_d")
plt.title("Average price (in Dollars) of each property ")
```

```python
# Average Rating of each property type

k = listing_merge.groupby(['property_type'])['review_scores_rating'].mean().sort_values(ascending=False).to_frame()
k=k[k.review_scores_rating.notnull()].reset_index()
k
```

```python
# How many property lies in each cancellation policy.

l = listing_merge.groupby(['cancellation_policy'])['property_type'].count().reset_index()
l
```

```python

explode = (0,0,0,0,0,0)

plt.pie(l['property_type'], labels = l['cancellation_policy'], autopct = '%.2f%%', pctdistance = 0.8,explode=explode)

centre_circle = plt.Circle((0,0),0.50, fc = 'white')

# gcf = get current figure
# gca = get current axis

fig = plt.gcf()

fig.gca().add_artist(centre_circle)

plt.axis('equal')
plt.title("Cancellation Policy")

plt.show()
```

```python
sns.barplot("cancellation_policy", y="property_type", data=l)
plt.title("property lies in each cancellation policy")
```
