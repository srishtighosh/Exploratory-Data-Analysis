# Problem 4: Sector Analysis 1

```python
#1.Extract the primary sector of each category list from the category_list column.
import re
ls=master_frame['category_list']
#ls.head()
ls1=[]
ls2=[]
for i in ls:
    ls1.append(re.findall('[a-zA-Z0-9\s-]+',str(i)))
#ls1
for j in range(len(ls1)):
    ls2.append(ls1[j][0])
#ls2
primary_sector=pd.DataFrame(ls2,columns=['category_list'])
primary_sector.head()
```

```python
mapping=pd.read_csv('C:/Users/Administrator/Desktop/Python/Pandas/Teclov/mapping.csv',encoding='ANSI')
mapping.head()
```

```python
#2.Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also 
#considered one of the main sectors)

mapping_1=mapping.melt(id_vars='category_list',var_name='Main_Sector')
mapping_1_sort=mapping_1[['category_list','Main_Sector']][mapping_1.value==1].sort_values('category_list')
merge_frame=pd.merge(mapping_1_sort,primary_sector,left_on='category_list',right_on='category_list')
merge_frame=merge_frame.drop_duplicates()
merge_frame.head()
```

```python
merge_frame.rename(columns={'category_list': 'Primary_Sector'},inplace=True)
merge_frame.head()
```
